```python
np.random.seed(22)
df = pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2021-07-18',periods=10),columns=['a','b','c','d'])
```

```python
df['new_col_square_d'] = df['d'].apply(lambda x: x **2)
```

```python
df
```

```python
df['new_col_square_d_log'] = df['d'].apply(lambda x:np.log(x) )
```

```python
df
```

```python
# log function did not return the log value of negative function
```

```python
def test(x):
    if (x<0):
        return x
    else:
        return np.log(x)
```

```python
df['new_col_square_d_log_1'] = df['d'].apply(test) 
```

```python
df
```

```python
def test1(x):
    if (x<0):
        return 1
    else:
        return 0
```

```python
df['new_col_a'] = df['a'].apply(test1)
```
