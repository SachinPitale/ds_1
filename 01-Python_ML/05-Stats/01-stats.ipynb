```python
## First we will check whether a distribution is Guassian distribution or not
## by using QQ plot (Quantile Quantile plot)
```

```python
import numpy as np
```

```python
np.random.normal(0,1,100)
```

```python
import numpy as np
import statsmodels.api as sm
import pylab as py

data_points = np.random.normal(0,1,100)
sm.qqplot(data_points, line='45')
py.show()
```

```python
import numpy as np

gfg = np.random.lognormal(0,1,100)
print(gfg)
```

```python
import seaborn as sns
sns.displot(gfg, kind="kde")
```

```python
np.log(gfg)
```

```python
sm.qqplot(np.log(gfg), line='45')
py.show()
```
