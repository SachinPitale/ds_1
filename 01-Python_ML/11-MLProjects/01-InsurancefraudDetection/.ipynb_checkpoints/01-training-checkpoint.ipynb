```python
from sklearn_pandas import CategoricalImputer
```

```python
imputer = CategoricalImputer()
```

```python
useful_data[['collision_type']]
```

```python
useful_data['collision_type'] =imputer.fit_transform(useful_data['collision_type']) 
```

```python
useful_data['collision_type'] 
```

```python
useful_data['property_damage'] = imputer.fit_transform(useful_data['property_damage'] )
```

```python
useful_data['police_report_available'] = imputer.fit_transform(useful_data['police_report_available'])
```

```python
useful_data['police_report_available'] = useful_data['police_report_available'].fillna(useful_data['police_report_available'].mode())
```

```python
useful_data.isnull().sum()
```

```python
cat_df= useful_data.select_dtypes(include=['object']).copy()
```

```python
cat_df
```

```python
set(cat_df['policy_csl'])
```

```python
cat_df['policy_csl'] = cat_df['policy_csl'].map({'100/300' : 1,'250/500' : 2.5 ,'500/1000': 5 })
```

```python
set(cat_df['insured_education_level'])
```

```python
cat_df['insured_education_level'] = cat_df['insured_education_level'].map({'JD':1, 'High School': 2, 'College':3, 'Masters':4, 'Associate': 5, 'MD': 6, 'PhD': 7 })
```

```python
set(cat_df['incident_severity'])
```

```python
cat_df['incident_severity'] = cat_df['incident_severity'].map({'Trivial Damage': 1, 'Minor Damage': 2, 'Major Damage': 3, 'Total Loss': 4})
```

```python
cat_df['insured_sex'] = cat_df['insured_sex'].map({'FEMALE': 0, 'MALE': 1})
```
