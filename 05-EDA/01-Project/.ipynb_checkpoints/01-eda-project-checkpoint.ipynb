## Step 2. Look at missing data

#### Calculate percentage of null value

```python
missing_number=df.isna().sum().sort_values(ascending=False)
```

```python
len(df)
```

```python
missing_percentage = missing_number/len(df)
```

```python
missing_info = pd.concat([missing_number,missing_percentage],axis=1,keys=['missing_number','missing_percentage'])
```

## Observation
We can see 19 columns have missing records and 6 columns have more than 15% of data missing.

I would remove columns with more than 15% missing records. However, that also depend on how much data fields are available and how important is the fields in predicting the final.

## find out releationship between all missing columns and target column

```python
len ([feature for feature in df.columns if df[feature].isna().sum() >=1])
```

```python
missing_feature = [feature for feature in df.columns if df[feature].isna().sum() >=1]
```

```python
missing_feature
```

```python
for feature in missing_feature:
    print(feature, np.round(df[feature].isna().sum()))
```

### find out releationship between all missing columns and target column
###  let's plot some amazing diagram



```python
df1=df.copy()
for feature in missing_feature:
    # let's make a variable that indicate 1 if the observation was missing or 0 otherwise
    df1[feature] = np.where(df1[feature].isnull(),1,0)
    
    
    # let's calculate the mean of target variable where the information is missing or present
    df1.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

#### Remove the top 6 missing columns (6 columns with more than 15% missing)

```python
missing_info.head(6).index
```

```python
df=df.drop(columns=missing_info.head(6).index)
```

```python
df.shape
```

##  Step 3. Split the dataset into numerical and categorical(String)

```python
numerics = ['int64','float64']
```

```python
df_num=df.select_dtypes(include=numerics)
```

```python
df_num.info()
```

```python
df_num.isna().sum()
```

```python
##  Replacing numerical data to median
```

```python
for num in df_num.columns:
    df_num[num].fillna(df_num[num].median(), inplace=True)
```

```python
df_cat=df.select_dtypes(include='object')
```

```python
df_cat
```

```python
df_cat.isna().sum()
```

```python
for i in df_cat.columns:
    df_cat[i].fillna(df_cat[i].mode()[0], inplace=True)
```

```python
df_cat.isna().sum()
```
