#### Lets compare number of goals scored per season. We will create a new column which would be the addition of goals scored by the home team and goals scored by the away team.

```python
df_season['Goal']=df_season['FTHG'] + df_season['FTAG'] 
```

```python
df_season.head()
```

```python
sns.barplot('Goal','Season', data=df_season)
plt.title("Number of Goal scored per Season")
plt.ylabel('EPL Season')
plt.xlabel('Goal Count')
plt.show()
```

```python
EPL_melted = df_season[['Season', 'FTHG', 'FTAG']]
EPL_melt = df_season.melt("Season",var_name="H/A",value_name="Goals")
EPL_melt
```

```python
sns.barplot('Goals','Season', hue='H/A', data=EPL_melt)
plt.title('Number of Goals Scored Per Season') 
plt.ylabel('Season')
plt.legend(['Home Goals', 'Away Goals'])
```

```python
df.head()
```

```python
EPL_heat = df.groupby(['Season','month']).sum().reset_index()
EPL_heat['Goals'] = EPL_heat ['FTHG'] + EPL_heat ['FTAG']
Final = EPL_heat.pivot('Season','month','Goals')
Final.head()
```

#### Now we plot a heatmap by passing the dataset to the heatmap() function provided by seaborn. annot argument can be set to True if you want to show the values in the blocks. Try it out!

```python
sns.heatmap(Final,annot=True, cmap='Blues',fmt='g')
plt.title('Goals Distribution based on month')
```

```python
df.head()
```

```python
EPL_count = df.groupby(['Season','month']).count().reset_index()
EPL_heat['G/M'] = round(EPL_heat['Goals'] / EPL_count['FTHG'])
Final = EPL_heat.pivot('Season', 'month', 'G/M')
Final
```

```python
sns.heatmap(Final, annot = True, cmap = 'Blues',fmt='g')
plt.title('Goals per match based on month') 
```
